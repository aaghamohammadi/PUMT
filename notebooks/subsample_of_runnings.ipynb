```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
dPath = Path("../docs/dumps")
```

```python
import pickle
```

```python
with open(dPath / "train_without_running.pkl", 'rb') as filename:
    X_train,y_train = pickle.load(filename)
```

```python
df = pd.concat([X_train,y_train],axis=1)
```

```python
df.head()
```

```python
mutants_not_running = df[df.numExecuted == 0].sample(n=50000)
```

```python
with open(dPath / "not_running.pkl", 'wb') as filename:
    pickle.dump(mutants_not_running, filename)
```
